## ğŸ§ª Interactive Learning: Let's See AI in Action!

**Time for your first AI experience!** Let's demonstrate the difference between traditional and AI vision:

````python
import subprocess
import time
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets

print("ğŸ¤– AI CAMERA DEMONSTRATION")
print("=" * 30)
print("\nğŸ¯ Let's see the difference between regular camera and AI camera!")
print("\nFirst, let's check if your IMX500 camera is properly connected...")

# Check camera detection
def check_camera_connection():
    """Check if IMX500 camera is detected"""
    try:
        result = subprocess.run(['rpicam-hello', '--list-cameras'], 
                              capture_output=True, text=True, timeout=10)
        
        if result.returncode == 0 and 'imx500' in result.stdout.lower():
            print("âœ… IMX500 AI Camera detected successfully!")
            print("\nğŸ“‹ Camera Information:")
            # Parse and display camera info
            lines = result.stdout.strip().split('\n')
            for line in lines:
                if 'imx500' in line.lower() or 'Available cameras' in line:
                    print(f"   {line}")
            return True
        else:
            print("âŒ IMX500 camera not detected")
            print("\nğŸ”§ Troubleshooting:")
            print("   â€¢ Check camera cable connection")
            print("   â€¢ Ensure camera is enabled: sudo raspi-config")
            print("   â€¢ Try: rpicam-hello --list-cameras")
            return False
            
    except subprocess.TimeoutExpired:
        print("â±ï¸ Camera detection timeout - this might indicate connection issues")
        return False
    except Exception as e:
        print(f"âŒ Error checking camera: {e}")
        return False

# Test camera connection
camera_ok = check_camera_connection()

if camera_ok:
    print("\nğŸ‰ Perfect! Your AI camera is ready for the demonstration.")
    print("\nğŸš€ Ready to proceed to the next notebook: Camera_Preview.ipynb")
else:
    print("\nğŸ”§ Please fix camera connection before proceeding.")
    print("\nğŸ“š Once camera is working, continue to Camera_Preview.ipynb")

print("\nğŸ’¡ What you'll see in the camera preview:")
print("   ğŸ“· Regular camera view: Just pixels and colors")
print("   ğŸ§  AI-enhanced view: Objects identified with bounding boxes and labels")
print("   ğŸ“Š Confidence scores: How sure the AI is about each detection")
````
