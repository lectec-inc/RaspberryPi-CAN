## Section 2: ADAS Features Deep Dive

Advanced Driver Assistance Systems (ADAS) are a collection of features designed to improve safety and comfort. Below is a list of common features.

### Common ADAS Features:
- **Forward Collision Warning (FCW):** Alerts driver to a potential crash.
- **Automatic Emergency Braking (AEB):** Automatically brakes to prevent a crash.
- **Lane Departure Warning (LDW):** Alerts driver when drifting out of a lane.
- **Lane Keeping Assist (LKA):** Actively steers to keep the vehicle in the lane.
- **Adaptive Cruise Control (ACC):** Maintains speed and a set following distance.
- **Blind Spot Detection (BSD):** Warns of vehicles in the driver's blind spot.
- **Rear Cross-Traffic Alert:** Warns of approaching vehicles when backing up.
- **Traffic Sign Recognition:** Reads speed limit and other signs.
- **Automatic High Beams:** Toggles high beams automatically based on traffic.

## Section 3: SAE Autonomy Levels
The Society of Automotive Engineers (SAE) defines 6 levels of driving automation, from Level 0 (no automation) to Level 5 (full automation).

### The 6 Levels:
- **Level 0 (No Automation):** The human driver performs all driving tasks.
- **Level 1 (Driver Assistance):** A single automated system assists the driver (e.g., ACC).
- **Level 2 (Partial Automation):** At least two automated systems work together (e.g., ACC + LKA). **The driver must remain fully engaged.** Most modern systems (Tesla Autopilot, Ford BlueCruise) are Level 2.
- **Level 3 (Conditional Automation):** The car can drive itself under specific conditions, but the driver must be ready to take back control at any time. The car monitors the driving environment.
- **Level 4 (High Automation):** The car can fully drive itself within a limited, mapped area (a 'geofence'). No driver attention is needed within this area. (e.g., Waymo robotaxis in Phoenix).
- **Level 5 (Full Automation):** The car can drive itself on any road, anywhere, under any conditions without human intervention. This does not exist yet.
