## Section 3: Color-Coded Dashboard

Now let's make it look professional using HTML and apply colors for safety warnings.

````python
# 3.1 - Color function for temperature
from IPython.display import display, HTML

def get_temp_color(temp):
    if temp is None:
        return 'grey'
    if temp > 60:
        return 'red' # Danger
    if temp > 45:
        return 'orange' # Warning
    return 'lime' # Nominal
````

````python
# 3.2 - HTML Dashboard Loop
try:
    for i in range(20):
        clear_output(wait=True)
        
        # Get data
        voltage = vesc.get_input_voltage()
        temp = vesc.get_fet_temperature()
        rpm = vesc.get_rpm()
        
        # Get color based on threshold
        temp_color = get_temp_color(temp)
        
        # Format data (or show 'N/A' if None)
        voltage_str = f'{voltage:.1f}V' if voltage is not None else 'N/A'
        temp_str = f'{temp:.1f}Â°C' if temp is not None else 'N/A'
        rpm_str = f'{int(rpm)}' if rpm is not None else 'N/A'
        
        # Create HTML string
        html_out = f"""
        <div style="border: 2px solid #44B6E5; padding: 10px; font-family: 'Space Grotesk', monospace; background-color: #f0f0f0;">
            <h2 style="margin: 0; padding: 0;">LIVE VEHICLE DASHBOARD</h2>
            <div style="display: flex; justify-content: space-around; font-size: 24px; text-align: center;">
                <p><strong>VOLTAGE:</strong><br>{voltage_str}</p>
                <p style="color: {temp_color};"><strong>TEMP:</strong><br>{temp_str}</p>
                <p><strong>RPM:</strong><br>{rpm_str}</p>
            </div>
        </div>
        """
        
        display(HTML(html_out))
        time.sleep(0.5)
except NameError:
    print("Connection not established. Please run the connection cell above first.")
````

## Section 4: Graphing

A dashboard is great for current status, but a graph is better for seeing trends. We will use `matplotlib` to plot data over time.

````python
# 4.1 - Collect data for 10 seconds and plot
import matplotlib.pyplot as plt

rpm_data = []
time_data = []
start_time = time.time()

print("Collecting data for 10 seconds... (spin the wheel by hand!)")
try:
    while time.time() - start_time < 10:
        rpm = vesc.get_rpm()
        if rpm is not None:
            rpm_data.append(rpm)
            time_data.append(time.time() - start_time)
        time.sleep(0.1)

    # Plotting the data
    plt.figure(figsize=(10, 5))
    plt.plot(time_data, rpm_data, marker='o', linestyle='-', color='#44B6E5')
    plt.title('Motor RPM over Time', fontname='Space Grotesk', fontsize=16)
    plt.xlabel('Time (s)', fontname='Space Grotesk')
    plt.ylabel('RPM', fontname='Space Grotesk')
    plt.grid(True, linestyle='--')
    plt.show()
except NameError:
    print("Connection not established. Please run the connection cell above first.")
````
