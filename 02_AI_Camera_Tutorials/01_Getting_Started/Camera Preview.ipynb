```python
import subprocess
import cv2
import numpy as np
from IPython.display import display, clear_output, Image
import ipywidgets as widgets
import threading
import time

print("ğŸ¥ AI CAMERA INITIALIZATION")
print("=" * 30)
print("\nğŸ”§ Preparing your IMX500 camera for preview...")

# â”€â”€ 0) Kill any stray libcamera processes & unload bcm2835_v4l2 so libcamera has exclusive access
print("ğŸ§¹ Cleaning up any existing camera processes...")
for p in ("libcamera-vid","libcamera-still","libcamera-raw"):
    subprocess.run(["pkill","-9",p], stderr=subprocess.DEVNULL)
subprocess.run(["sudo","rmmod","bcm2835_v4l2"], stderr=subprocess.DEVNULL)

print("âœ… Camera resources prepared")
print("\nğŸ® Interactive camera preview controls ready!")
print("ğŸ’¡ Click the button below to start/stop the preview")
```

```python
# Stop button in notebook - Updated for headless operation
stopButton = widgets.ToggleButton(
    value=False,
    description='â–¶ï¸ Start Preview',
    button_style='success',
    icon='play',
    layout=widgets.Layout(width='200px', height='40px')
)

# Status display
status_label = widgets.HTML(
    value="<b>ğŸ“· Camera Status:</b> Ready to start (Headless Mode)",
    layout=widgets.Layout(margin='10px 0px')
)

display(widgets.VBox([status_label, stopButton]))

def camera_preview():
    """Main camera preview function - optimized for headless operation"""
    # Update button appearance when running
    stopButton.description = 'â¹ï¸ Stop Preview'
    stopButton.button_style = 'danger'
    stopButton.icon = 'stop'
    status_label.value = "<b>ğŸ“· Camera Status:</b> <span style='color: green'>â— LIVE (Headless Mode)</span>"
    
    display_handle = display(None, display_id=True)
    frame_count = 0
    start_time = time.time()
    
    try:
        while not stopButton.value:
            # Capture still frames periodically for headless operation
            output_file = f"/tmp/preview_{frame_count}.jpg"
            
            cmd = [    
                "rpicam-still",
                "--width", "640",     # Moderate resolution for good performance
                "--height", "480",
                "--immediate",        # Take photo immediately
                "--timeout", "100",   # Quick timeout
                "-o", output_file
            ]
            
            result = subprocess.run(cmd, capture_output=True, text=True, timeout=5)
            
            if result.returncode == 0:
                try:
                    # Read and display the captured frame
                    with open(output_file, 'rb') as f:
                        img_data = f.read()
                    
                    display_handle.update(Image(data=img_data))
                    frame_count += 1
                    
                    # Update status every few frames
                    if frame_count % 5 == 0:  # Update every 5 frames
                        elapsed = time.time() - start_time
                        fps = frame_count / elapsed if elapsed > 0 else 0
                        status_label.value = f"<b>ğŸ“· Camera Status:</b> <span style='color: green'>â— LIVE (Headless)</span> | Photos: {frame_count} | Rate: {fps:.1f}/sec"
                    
                    # Clean up temporary file
                    try:
                        import os
                        os.remove(output_file)
                    except:
                        pass
                        
                except Exception as e:
                    status_label.value = f"<b>ğŸ“· Camera Status:</b> <span style='color: orange'>â— Warning - Display issue</span>"
                    
            else:
                status_label.value = f"<b>ğŸ“· Camera Status:</b> <span style='color: red'>â— ERROR - Camera capture failed</span>"
                break
            
            # Wait between captures (headless-friendly timing)
            time.sleep(1.5)  # ~0.7 FPS - good for headless learning
            
    except Exception as e:
        status_label.value = f"<b>ğŸ“· Camera Status:</b> <span style='color: red'>â— ERROR - {str(e)}</span>"
    finally:
        # Reset button appearance
        stopButton.description = 'â–¶ï¸ Start Preview'
        stopButton.button_style = 'success'
        stopButton.icon = 'play'
        status_label.value = "<b>ğŸ“· Camera Status:</b> Stopped"

def on_button_change(change):
    """Handle button state changes"""
    if change['new']:  # Button was pressed (start preview)
        # Start preview in a separate thread
        threading.Thread(target=camera_preview, daemon=True).start()
    # When button is released, the preview loop will detect stopButton.value=False and stop

stopButton.observe(on_button_change, names='value')
```
